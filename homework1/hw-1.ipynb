# Homework 1

### Environment Description 

```python
import numpy as np

P = np.array([[0.4, 0.4, 0.2, 0.0],
              [0.2, 0.5, 0.2, 0.1],
              [0.1, 0.4, 0.4, 0.1],
              [0.0, 0.5, 0.1, 0.4]])

v0 = np.array([1, 0, 0, 0])
```

### State Distribution

1. Calculate the state distribution at t=2

```python
P2 = np.linalg.matrix_power(P, 2)

print(f"P^2 = {P2}")

v2 = v0.dot(P2)

print(f"State distribution at t=2: {v2}")
```

2. What is the probability that the System is in Blackout state (S4) at t=10?

```python
P10 = np.linalg.matrix_power(P, 10)

print(f"P^10 = {P10}")

v10 = v0.dot(P10)

print(f"State distribution at t=10: {v10}")

print(f"Probability of System being in S4 at t=10: {v10[3]}")
```

3. Is there a stationary distribution?

```python
# Initial Case

t = 100000

v = v0.dot(np.linalg.matrix_power(P, t))

print(f"State distribution: {v}")
```
