```python
import numpy as np

gamma = 0.9

Pb = np.array([
    [0.5, 0.5, 0.0, 0.0],
    [0.5, 0.5, 0.0, 0.0],
    [0.0, 0.0, 0.5, 0.5],
    [0.25, 0.25, 0.25, 0.25]
])

Pg = np.array([
    [0.0, 0.0, 0.5, 0.5],
    [0.0, 0.0, 0.5, 0.5],
    [0.0, 0.0, 0.5, 0.5],
    [0.0, 0.0, 0.5, 0.5]
])

Cb = np.array([1, 5, 0, 1])
Cg = np.array([2, 2, 2, 2])

# pi = [B, G, B, B]
policy_actions = [0, 1, 0, 0] # 0 for B, 1 for G

# --- 1 - POLICY EVALUATION ---
P_pi = np.zeros((4, 4))
c_pi = np.zeros(4)

for s in range(4):
    if policy_actions[s] == 0: # Action B
        P_pi[s] = Pb[s]
        c_pi[s] = Cb[s]
    else: # Action G
        P_pi[s] = Pg[s]
        c_pi[s] = Cg[s]

# Solve J = c_pi + gamma * P_pi * J => (I - gamma*P_pi)J = c_pi
I = np.eye(4)
J_pi = np.linalg.solve(I - gamma * P_pi, c_pi)

# --- 2 - POLICY IMPROVEMENT ---
Q = np.zeros((4, 2))
for s in range(4):
    Q[s, 0] = Cb[s] + gamma * np.dot(Pb[s], J_pi) # Value for B
    Q[s, 1] = Cg[s] + gamma * np.dot(Pg[s], J_pi) # Value for G

best_actions = np.argmin(Q, axis=1)

print("--- VALORES INTERMÉDIOS ---")
print(f"Vetor de Custo c_pi: {c_pi}")
print(f"Custo-to-go J^pi: {np.round(J_pi, 4)}")
print("\nTabela de Valores Q (Q(s, B) | Q(s, G)):")
for s, name in enumerate(["LS", "LC", "HS", "HC"]):
    print(f"{name}: {np.round(Q[s, 0], 4)} | {np.round(Q[s, 1], 4)}")

print("\n--- CONCLUSÃO ---")
optimal = np.array_equal(best_actions, policy_actions)
print(f"A política intuitiva é ótima? {'Sim' if optimal else 'Não'}")
if not optimal:
    print(f"Melhor política encontrada: {['B' if a==0 else 'G' for a in best_actions]}")
```
